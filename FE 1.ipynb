# Non Normal dist

```python
df=pd.read_csv(r"C:\Python\Data Science sid\EDA\placement.csv")
df
```

```python
# q1=25%
# q3=75%
# iqr=q3-q1
# upper_fense=q3+1.5*(iqr)
# upper_fense=q3-1.5*(iqr)
```

```python
q1=df["placement_exam_marks"].quantile(0.25)
q3=df["placement_exam_marks"].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
upper_fence=q3+1.5*(iqr)
lower_fence=q3-1.5*(iqr)
```

```python
new1_df= df[(df["placement_exam_marks"]<upper_fence) & (df["placement_exam_marks"]>lower_fence)]
new1_df
```
